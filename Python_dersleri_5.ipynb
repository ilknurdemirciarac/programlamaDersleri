```python
#Kullanıcıdan alınan isimleri baş harfi büyük ise listeye alan değilse kullanıcıya uyarı vererek devame den programı yazınız
isimler=[]
while True:
  isim=input("İsim giriniz : ")
  if not isim:
    break
  if isim[0].islower():
    print("Lütfen isminizin ilk harfini büyük yazınız. Tekrar deneyin")
    continue
  isimler.append(isim)
print(isimler)
```

```python
""" Bir admin panelini tasarlarken aşağıda belirtilen şartlarda kullanıcıyı bilgilendirerek bir program yazalım
kullanıcı adı ve şifresiin isteyin
kullancıı adı doğru şifre yanlış
şifre doğru kullancıı adı yanlış
kullanıcıadı ve şifre yanlış
kullanıcı adı ve şifre boş
kullanıcı adı ve şifre doğru"""
kAd="admin"
sfr="1234"
while True:
  kullaniciAdi=input("Kullanıcı adını giriniz")
  sifre=input("Şifrenizi giriniz")
  if not kullaniciAdi or not sifre:
    print("Kullanıcı adı veya şifre boş bırakılamaz")
    continue
  elif kAd!=kullaniciAdi and sfr!=sifre:
    print("Kullanıcı ad ve şifre hatalı")
    continue
  elif kAd==kullaniciAdi and sfr!=sifre:
    print("Kullanıcı adını doğru girdiniz Fakat şifre hatalı")
    continue
  elif kAd!=kullaniciAdi and sfr==sifre:
    print("Kullanıcı adını yanlış girdiniz Fakat şifre başarılı")
    continue
  else:
    break
print("Sisteme giriş başarılı")

```

```python
#Klavyeden girilen sayı 0 sıfır olana kadar girilen sayılardan çift sayıları toplayan tek sayıları çıkaran ve sonucu ekrana yazdıran program
print("Sonucu ekranda görmek için 0(sıfır) tuşuna basınız")
toplam=0
while True:
  sayi=int(input("Bir sayı giriniz"))
  if sayi==0:
    break
  elif sayi%2==0:
    toplam+=sayi
  else:
    toplam-=sayi
print("Sayıların toplamı : ", toplam)
```
